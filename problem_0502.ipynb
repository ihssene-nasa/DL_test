2). Get rid of the columns that have more than 97% missing values.

```python
nan_cols = set(df.columns[df.isnull().mean() > 0.97])
```

```python
df.drop(nan_cols, axis = 1)
```

3). Get rid of the unnecessary columns. HINT: Those columns with "time" in the name and those that are obviously unnecessary.

```python
X = ['Unnamed: 0','raw_timestamp_part_2','raw_timestamp_part_2','cvtd_timestamp']
df = df.drop(X, axis = 1)
df = df.drop(nan_cols, axis = 1)
df.head()
```

```python
#Select categorical columns
categ_cols = df.columns
n = df._get_numeric_data().columns
list(set(categ_cols) - set(n))
```

4). Label encode the responde variable. HINT: use preprocessing.LabelEncoder(). 

```python
from sklearn import preprocessing
Encoder = preprocessing.LabelEncoder()
df['classe']= Encoder.fit_transform(df['classe'])
df['new_window']= Encoder.fit_transform(df['new_window'])
df['user_name']= Encoder.fit_transform(df['user_name'])
```

```python
df.head()
```

5). Carry out min-max scaling of the exploratory variables. HINT: use preprocessing.MinMaxScaler().

```python
from sklearn import preprocessing
for col in df.columns : 
    x = df[col].values.reshape(-1, 1) 
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df[col] = x_scaled
```

6). Choose an algorithm and carry out the predictive analysis.

- Optimize the hyperparameter(s)
- Calculate the accuracy.
- Aim for upper 90% accuracy.

```python
corr = df.corr()
corr['classe'].sort_values(ascending=False)
```

```python
x=df[["pitch_forearm", "magnet_arm_x",'accel_arm_x', "total_accel_forearm", "magnet_dumbbell_z", "accel_dumbbell_x","roll_arm","pitch_dumbbell","total_accel_belt","gyros_dumbbell_x"]]

y= df["classe"].astype(int)

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)

#applying tree algorithm
clf=RandomForestClassifier(n_estimators=500)  #Creating a random forest with 100 decision trees
clf.fit(x_train, y_train)  #Training our model
predictions=clf.predict(x_test)  #testing our model
print("Accuracy:", metrics.accuracy_score(y_test, predictions))  #Measuring the accuracy of our model
```

```python
from xgboost import XGBRegressor
XGB = XGBClassifier(n_estimators=500)
XGB.fit(x_train, y_train)
predictions = XGB.predict(x_test)
print("Accuracy:", metrics.accuracy_score(y_test, predictions))  
```
